```python
Y = mezcla['resultado']
mezcla = mezcla.drop('resultado',axis=1)
```

```python

matriz_ref = [["L00","U00","B02"],
              ["B00", "U02", "R02"],
              ["F00", "U20", "L02"],
              ["R00", "U22", "F02"],
              ["B22", "D20", "L20"],
              ["R22", "D22", "B20"],
              ["F20", "D00", "L22"],
              ["F22", "D02", "R20"]]

# Crear una lista vacía para almacenar los valores condicionados
valores_condicionado = []
	
for i in range(1000000):
    almacenamiento_provisional = [0, 0, 0, 0, 0, 0, 0, 0]

    for j in range(8):
        if mezcla.iloc[i][matriz_ref[j][0]] % 4 == 0:
            almacenamiento_provisional[j] = -1
        elif mezcla.iloc[i][matriz_ref[j][1]] % 4 == 0:
            almacenamiento_provisional[j] = 0
        else:
            almacenamiento_provisional[j] = 1

    valores_condicionado.append(almacenamiento_provisional)
    if(i % 1000 == 0):
        print(i)
# Crear el DataFrame directamente desde la lista de valores condicionados
valores_condicionado = pd.DataFrame(valores_condicionado, columns=['P000', 'P200', 'P020', 'P220', 'P002', 'P202', 'P022', 'P222'])

print(valores_condicionado)
```
